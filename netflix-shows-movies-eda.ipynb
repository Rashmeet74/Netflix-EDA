Netflix Shows & Movies Analysis-

This dataset contains details about the movies and tv shows. Details such as the Cast,Title,Director,Release year,Rating,Duration etc.The objective of this analysis is to uncover insights on genres,rating and trends.

**#Load and clean the data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/kaggle/input/netflix-shows-and-movies/netflix_titles_nov_2019.csv') # Load data


df = df.dropna(subset=['title']) # Drop rows missing title


df = df.copy()

df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available') # Clean director/cast fields


df = df.copy()
df['country'] = df['country'].dropna().str.split(',').str[0]# Simplify country to first listed


df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year # Parse dates



```

## Key Questions:
1. **Movies VS TV Shows** – which is more common?  
2. **Top genres** – what content dominates Netflix?  
3. **Country distribution** – which countries produce the most?  
4. **Growth over time** – how has Netflix content grown yearly?


 Q1: Movies vs TV Shows

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution: Movies vs TV Shows')
plt.xlabel('')
plt.show()
```

🔸 Q2: Top genres – what content dominates Netflix?

```python
plt.figure(figsize=(8,5))
top_genres = df['listed_in'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('Top 10 Genres on Netflix')
plt.show()
```

Q3: Top Countries- which countries produce the most?

```python
plt.figure(figsize=(8,5))
countries = df['country'].value_counts().head(10)
sns.barplot(x=countries.values, y=countries.index, palette='magma')
plt.title('Top 10 Countries by Titles')
plt.show()
```

 Q4: Yearly Content Growth

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Count titles per year
year_counts = df['year_added'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 4))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title('Content Added Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Conclusion-

**#Key Insights:**
- Netflix has more **Movies** than **TV Shows**
- Top genres include International TV, Dramas, and Comedies
- The **USA** leads in content production
- There was rapid content growth from **2017–2020**, then slowed
